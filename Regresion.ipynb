# Linear Regression

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn import linear_model
from sklearn.metrics import mean_absolute_error, mean_squared_error, median_absolute_error, r2_score, explained_variance_score
```

```python
# Load the data
data_a = pd.read_csv("df_regresion_lineal_1.cvs")
data_b = pd.read_csv("df_regresion_lineal_2.cvs")
data_c = pd.read_csv("df_regresion_lineal_3.cvs")
data_d = pd.read_csv("df_regresion_lineal_pop.cvs")

# Create different models for each dataset
model_a = linear_model.LinearRegression()
model_b = linear_model.LinearRegression()
model_c = linear_model.LinearRegression()
model_d = linear_model.LinearRegression()

# We asign a different model to each dataset 

# first model
x_a = np.asanyarray(data_a[['x']])
y_a = np.asanyarray(data_a[['y']])
model_a.fit(x_a,y_a)
# Generating predicted values (yp) and print the errors
yp_a = model_a.predict(x_a); print("Model 1"); 
e1 = mean_absolute_error(y_a,yp_a); print("Average absolute error: ", e1)
e2 = mean_squared_error(y_a,yp_a); print("Avergae square error: ", e2)
e3 = median_absolute_error(y_a,yp_a); print("Median absolute error: ", e3)
e4 = r2_score(y_a,yp_a); print(" R^2 Coeficient: ", e4)
e5 = explained_variance_score(y_a,yp_a); print("Variance coeficient: ", e5)
print("")

# second model
x_b = np.asanyarray(data_b[['x']])
y_b = np.asanyarray(data_b[['y']])
model_b.fit(x_b,y_b)
# Generating predicted values (yp) and print the errors
yp_b = model_b.predict(x_b); print("Model 1"); 
e1 = mean_absolute_error(y_b,yp_b); print("Average absolute error: ", e1)
e2 = mean_squared_error(y_b,yp_b); print("Avergae square error: ", e2)
e3 = median_absolute_error(y_b,yp_b); print("Median absolute error: ", e3)
e4 = r2_score(y_b,yp_b); print(" R^2 Coeficient: ", e4)
e5 = explained_variance_score(y_b,yp_b); print("Variance coeficient: ", e5)
print("")

# third model
x_c = np.asanyarray(data_c[['x']])
y_c = np.asanyarray(data_c[['y']])
model_c.fit(x_c,y_c)
# Generating predicted values (yp) and print the errors
yp_c = model_c.predict(x_c); print("Model 3"); 
e1 = mean_absolute_error(y_c,yp_c); print("Average absolute error: ", e1)
e2 = mean_squared_error(y_c,yp_c); print("Avergae square error: ", e2)
e3 = median_absolute_error(y_c,yp_c); print("Median absolute error: ", e3)
e4 = r2_score(y_c,yp_c); print(" R^2 Coeficient: ", e4)
e5 = explained_variance_score(y_c,yp_c); print("Variance coeficient: ", e5)
print("")

# Fourth model
x_d = np.asanyarray(data_d[['year']])
y_d = np.asanyarray(data_d[['pop']])
model_d.fit(x_d,y_d)
# Generating predicted values (yp) and print the errors
yp_d = model_d.predict(x_d); print("Model 4"); 
e1 = mean_absolute_error(y_d,yp_d); print("Average absolute error: ", e1)
e2 = mean_squared_error(y_d,yp_d); print("Avergae square error: ", e2)
e3 = median_absolute_error(y_d,yp_d); print("Median absolute error: ", e3)
e4 = r2_score(y_d,yp_d); print(" R^2 Coeficient: ", e4)
e5 = explained_variance_score(y_d,yp_d); print("Variance coeficient: ", e5)
print("")


# Graph each model with itÂ´s dataset to se the accuracy


plt.figure(); plt. grid(); plt.title("Linear Regression 1")
plt.xlabel("X"); plt.ylabel("Y")
#Graph the data with blue circles and the model predictions with a red line
plt.plot(x_a,y_a,'bo')
plt.plot(x_a,yp_a,'r-', lw=2)
plt.show()


plt.figure(); plt. grid(); plt.title("Linear Regression 2")
plt.xlabel("X"); plt.ylabel("Y")
#Graph the data with blue circles and the model predictions with a red line
plt.plot(x_b,y_b,'bo')
plt.plot(x_b,yp_b,'r-', lw=2)
plt.show()

plt.figure(); plt. grid(); plt.title("Linear Regression 3")
plt.xlabel("X"); plt.ylabel("Y")
#Graph the data with blue circles and the model predictions with a red line
plt.plot(x_c,y_c,'bo')
plt.plot(x_c,yp_c,'r-', lw=2)
plt.show()

plt.figure(); plt. grid(); plt.title("Linear Regression 4")
plt.xlabel("Year"); plt.ylabel("Population")
#Graph the data with blue circles and the model predictions with a red line
plt.plot(x_d,y_d,'bo')
plt.plot(x_d,yp_d,'r-',lw=2)
plt.show()
```
